# Demo notebook: calculate $\pi$ using monte carlo methods
In this notebook, we demonstrate the main techniques of Jupyter

```python
import random
```

```python
N = 1000
points = [ ]
```

```python
hits = 0
for i in range(N):
    x, y = random.random(), random.random()
    if x**2 + y**2 < 1:
        hits += 1
        points.append((x,y, True))
    else:
        points.append((x,y, False))        
```

```python
%matplotlib inline
from matplotlib import pyplot
x, y, colors = zip(*points)
pyplot.scatter(x, y, c=colors)
```

# Our formula
* square area = $(2 r)^2$
* circle area = $\pi r^2$
* circle / square = $\pi r^2 / 4 r^2$ = $\pi / 4$
* $\pi = (circle/square) * 4$

```python
fraction = hits / N
4 * fraction
```
